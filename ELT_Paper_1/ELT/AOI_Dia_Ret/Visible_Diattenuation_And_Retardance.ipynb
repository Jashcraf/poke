```python
import numpy as np
import sys
sys.path.insert(1,'C:/Users/UASAL-OPTICS/Desktop/poke/') # One day I'll actually package poke!
import poke.poke_core as pol
from astropy.io import fits
import astropy.units as u
# import poppy
import matplotlib.pyplot as plt
```

```python
pth = 'ELT_BareAg_DRAOI_445.fits'
data = fits.open(pth)[0].data
# shape is npix x npix x surf x diat x retardance
```

```python
surf = 1

for surf in range(data.shape[2]):

    plt.figure(figsize=[12,5])
    plt.suptitle('Surface {}'.format(surf+1))
    plt.subplot(131)
    plt.imshow(data[:,:,surf,0])
    plt.colorbar()
    plt.title('AOI [rad]')

    plt.subplot(132)
    plt.imshow(data[:,:,surf,1])
    plt.colorbar()
    plt.title('Diattenuation')

    plt.subplot(133)
    plt.imshow(data[:,:,surf,2]+np.pi)
    plt.colorbar()
    plt.title('Retardance [rad]')

    plt.show()
```

```python
plt.figure(figsize=[12,5])
plt.suptitle('Total')
plt.subplot(121)
plt.imshow(np.sum(data[:,:,:,1],axis=2))
plt.colorbar()
plt.title('Diattenuation')

plt.subplot(122)
plt.imshow(np.sum(data[:,:,:,2]+np.pi,axis=2))
plt.colorbar()
plt.title('Retardance [rad]')

plt.show()
```
